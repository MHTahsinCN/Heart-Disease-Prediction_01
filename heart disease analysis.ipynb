```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
```

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
type(over_50)
```

```python
type(heart_disease)
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-bright')

fig, (ax0, ax1) = plt.subplots(nrows = 2,
                             ncols = 1,
                             figsize = (10,10))

### Working with ax0
scatter = ax0.scatter(x=over_50["age"],
                     y= over_50["chol"],
                     c= over_50["target"])
##customize the plot:
# set limit
ax0.set_xlim([50, 80])
# set titles
ax0.set(title = "Heart Disease and Cholesterol Analysis",
       xlabel = 'Age',
       ylabel = 'Cholesterol')
# set legend
ax0.legend(*scatter.legend_elements(),
          title = "Target")
# set axline
ax0.axhline(y = over_50["chol"].mean(),
           linestyle = '--')


### working with ax1
scatter = ax1.scatter(x = over_50['age'],
                     y = over_50['thalach'],
                     c = over_50['target'])
# set limits
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])
# set titles
ax1.set(title = 'Heart Disease and Thalach Analysis',
       xlabel = 'Age',
       ylabel = 'Max heart rate')
# set legend
ax1.legend(*scatter.legend_elements(),
          title = 'Target')
# set axline
ax1.axhline(y = over_50['thalach'].mean(),
           linestyle = '--')

fig.suptitle("Heart  Disease Analysis",
            fontsize = 16, fontweight = 'bold');

plt.savefig("HDA")
```
